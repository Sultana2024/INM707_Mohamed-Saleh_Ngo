```python
import ale_py
import gymnasium as gym
import GPUtil
import numpy as np
import os
import pybullet
import ray
import torch
import shimmy

from ale_py import ALEInterface, roms
from gymnasium.spaces import Discrete, Box
from ray import rllib
from ray.rllib.algorithms.ppo import PPOConfig
from ray import train
from ray import tune
import pkg_resources
from pkg_resources._vendor.packaging.version import parse as parse_version
```

```python
if ray.is_initialized():
    # Shut down in case initialised
    ray.shutdown()
torch.cuda.empty_cache()

# Set environment variables before Ray initialisation
os.environ["CUDA_VISIBLE_DEVICES"] = "0"

# Check GPU availability
print(f"CUDA available: {torch.cuda.is_available()}")
if torch.cuda.is_available():
    print(f"GPU device: {torch.cuda.get_device_name(0)}")
    print(f"CUDA version: {torch.version.cuda}")
    print(f"number: {torch.cuda.device_count()}")
```

```python
ray.init(ignore_reinit_error=True, num_gpus=1)
```

```python
# Check resources
print(ray.get_gpu_ids())
print(ray.available_resources())
```

```python
#Sources:
#https://docs.ray.io/en/latest/rllib/rllib-env.html
#https://github.com/ray-project/ray/blob/master/rllib/tuned_examples/ppo/cartpole_ppo.py

from ray.tune.registry import register_env

def env_creator_cartpole(config):
    env = gym.make("CartPole-v1")
    env.observation_space = Box(low=-np.inf, high=np.inf, shape=(4,), dtype=np.float32)
    env.action_space = Discrete(2)  # Discrete action space (left or right)
    return env

# Register the CartPole environment
register_env("CartPole-v1", env_creator_cartpole)

config = PPOConfig()
config.environment("CartPole-v1")
config.resources(num_gpus=1)
config.framework("torch")
config['lr']=0.0003
config['clip_param']=0.2

algo = config.build()
```

```python
score = []
ppo_info = []
for i in range(200):
    result = algo.train()
    score.append(result['episode_reward_mean'])
    ppo_info.append(result['info'])
    print(f"Iteration: {i}, Reward: {result['episode_reward_mean']}")
```

```python
import pandas as pd
score_series = pd.Series(score)
score_series.to_csv('score.csv', index=False)
```

```python
# Plot Score per training iteration
import matplotlib.pyplot as plt
rolling = score_series.rolling(window=10).mean().rename("Rolling Average")
plt.plot(score)
plt.plot(rolling, label='Rolling Average')
plt.xlabel("Training Iteration")
plt.ylabel("Mean Episode Reward")
plt.title("PPO Performance – CartPole")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Load the saved model for evaluation
from ray.rllib.algorithms.ppo import PPO
algo = PPO(config={"env": "CartPole-v1", "framework": "torch"})
algo.restore("/path/to//tmp/tmprle27sda")
```

```python
ray.init(ignore_reinit_error=True, num_gpus=1, num_cpus=2)
```

```python
env = gym.make("CartPole-v1")
env.observation_space = gym.spaces.Box(low=-np.inf, high=np.inf, shape=(4,), dtype=np.float32)
env.action_space = gym.spaces.Discrete(2)

def evaluate_ppo_custom(algo, num_episodes=8):
    scores = []

    for i in range(num_episodes):
        state, info= env.reset()  # Reset the environment
        done = False
        score_ep = 0

        while not done:
            action = algo.compute_single_action(state, explore=False)
            next_state, reward, terminated, truncated, info= env.step(action)
            score_ep += reward
            state = next_state

        scores.append(score_ep)

    average_reward = np.mean(scores)
    print(f"Average Reward {num_episodes} episodes: {average_reward}")
    return average_reward, scores
```

```python
result, score = evaluate_ppo_custom(algo)
```
